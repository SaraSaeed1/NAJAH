# import libraries

```python
import pandas as pd                                               #read excel files (detaset)
import numpy as np                                                #deal with arrays 
from sklearn.model_selection import train_test_split              #split data
from sklearn.preprocessing import LabelEncoder                    #encode data 
%matplotlib inline
import matplotlib.pyplot as plt                                   #plot data
import seaborn as sns                                             #plot data
```

```python
#load dataset
data = pd.read_excel('Dataset.xlsx')
```

```python
#display data
data.head(n=10)
```

# Data PreProcessing 

```python
#delete unusful data
data = data.drop(['Timestamp'], axis=1)
data.head()
```

```python
#encode data (from string to numerical values )
data['نوع المشروع ؟'] = LabelEncoder().fit_transform(data['نوع المشروع ؟'])

data['رأس المال'] = LabelEncoder().fit_transform(data['رأس المال'])

data['هل يوجد منافسين ؟'] = LabelEncoder().fit_transform(data['هل يوجد منافسين ؟'])

data['هل مشروعك ؟'] = LabelEncoder().fit_transform(data['هل مشروعك ؟'])

data['عدد الموظفين المحتمل '] = LabelEncoder().fit_transform(data['عدد الموظفين المحتمل '])

data['هل مشروعك يوجد به شركاء ؟'] = LabelEncoder().fit_transform(data['هل مشروعك يوجد به شركاء ؟'])

data['هل يوجد خطط للمخاطر ؟'] = LabelEncoder().fit_transform(data['هل يوجد خطط للمخاطر ؟'])

data['كم المبلغ للمخاطر ؟'] = LabelEncoder().fit_transform(data['كم المبلغ للمخاطر ؟'])

data['هل يوجد ممول للمشروع ؟'] = LabelEncoder().fit_transform(data['هل يوجد ممول للمشروع ؟'])

data['هل تم اللجوء فقط الي التمويل ؟'] = LabelEncoder().fit_transform(data['هل تم اللجوء فقط الي التمويل ؟'])

data['هل مشروعك'] = LabelEncoder().fit_transform(data['هل مشروعك'])
```

```python
#display filnal dataset
data.head(n=10)
```

```python
#plot categories 
sns.countplot(data['هل مشروعك'])
plt.xlabel('Label')
plt.title('Number of each sentiment')
print('-------------------------------')
print(data['هل مشروعك'].value_counts())
print('-------------------------------')
```

```python
#correlation matrix
corrmat = data.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.20, square=True);
```

# split data

```python
#split the the target and the features 
X = data.drop('هل مشروعك',axis = 1 )
X = X.drop("نوع المشروع ؟",axis = 1 )
Y = data['هل مشروعك']
```

```python
#display X
X
```

```python
#display y
Y
```

```python
#Splitting data

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=44, shuffle =True)

#Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

# Building Model

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
Algorithm1 = LogisticRegression()
Algorithm1.fit(X_train, y_train)

#Calculating Details
print('Logistic Regression Model Train Score is : ' ,Algorithm1.score(X_train, y_train))
print('Logistic Regression Model Test Score is : ' , Algorithm1.score(X_test, y_test))
y_pred1 = Algorithm1.predict(X_test)
```

# Support Victor Machine Model

```python
from sklearn import svm
Algorithm2 = svm.SVC()
Algorithm2.fit(X_train, y_train)

#Calculating Details
print('Support Victor Machine Model Train Score is : ' ,Algorithm2.score(X_train, y_train))
print('Support Victor Machine Test Score is : ' , Algorithm2.score(X_test, y_test))
y_pred2 = Algorithm2.predict(X_test)
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
Algorithm3 = RandomForestClassifier(max_depth=2, random_state=0)
Algorithm3.fit(X_train, y_train)

#Calculating Details
print('RandomForestClassifier Train Score is : ' ,Algorithm3.score(X_train, y_train))
print('RandomForestClassifier Test Score is : ' , Algorithm3.score(X_test, y_test))
y_pred3 = Algorithm3.predict(X_test)
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor
Algorithm4 = KNeighborsRegressor(n_neighbors=2)
Algorithm4.fit(X_train, y_train)

#Calculating Details
print('KNN Model Train Score is : ' ,Algorithm4.score(X_train, y_train))
print('KNN Test Score is : ' , Algorithm4.score(X_test, y_test))
```

# Evaluation

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print("Evaluation of Logistic Regression Model")
print("---------------------------------------")
print("\n Accuracy score :")
print(accuracy_score(y_test,y_pred1))
print("\n confusion_matrix :")
print(confusion_matrix(y_test,y_pred1))
print("\n classification report :")
print(classification_report(y_test,y_pred1))

print("\n Evaluation of Support Victor Machine Model")
print("---------------------------------------------")
print("Accuracy score :")
print(accuracy_score(y_test,y_pred2))
print("\n confusion_matrix :")
print(confusion_matrix(y_test,y_pred2))
print("\n classification report :")
print(classification_report(y_test,y_pred2))

print("\n Evaluation of Random Forest Classifier Model")
print("-----------------------------------------------")
print("Accuracy score :")
print(accuracy_score(y_test,y_pred3))
print("\n confusion_matrix :")
print(confusion_matrix(y_test,y_pred3))
print("\n classification report :")
print(classification_report(y_test,y_pred3))
```
